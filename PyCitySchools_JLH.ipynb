```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Create Variables
school_count = school_data_complete['school_name'].nunique()
student_count = school_data_complete['Student ID'].nunique()
total_budget = school_data['budget'].sum()
avg_math = school_data_complete['math_score'].mean()
avg_reading = school_data_complete['reading_score'].mean()
passing_rate = (avg_math + avg_reading)/2
passing_math = school_data_complete.loc[school_data_complete['math_score'] >= 70]
passing_percent_math = ((passing_math['Student ID'].nunique())/student_count)*100
passing_reading = school_data_complete.loc[school_data_complete['reading_score'] >= 70]
passing_percent_reading = ((passing_reading['Student ID'].nunique())/student_count)*100

#Create Series
district_data = {'Total Schools': school_count,
                             'Total Students' : "{:,}".format(student_count),
                             'Total School Budget' : "${:,}".format(total_budget),
                             'Average Math Score' : "{:.2f}%".format(avg_math),
                             'Average Reading Score' : "{:.2f}%".format(avg_reading),
                             '% Passing Math' : "{:.2f}%".format(passing_percent_math),
                             '% Passing Reading' : "{:.2f}%".format(passing_percent_reading),
                             '% Overall Passing Rate': "{:.2f}%".format(((avg_math + avg_reading)/2))}

#Create DataFrame from Series
district_summary = pd.DataFrame(district_data, index = [''])

district_summary

```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
grouped_school_data_df = school_data_complete.groupby(['school_name'])

school_type = grouped_school_data_df['type'].first()
total_students = grouped_school_data_df['Student ID'].nunique()
school_budget = grouped_school_data_df['budget'].first()
avg_math = grouped_school_data_df['math_score'].mean()
avg_reading = grouped_school_data_df['reading_score'].mean()
Per_Student_Budget = (school_budget/total_students)

#filter and groupby school name math scores above 70
pass_math_data = school_data_complete.loc[school_data_complete["math_score"] >= 70, :]
pass_math_grouped = pass_math_data.groupby(['school_name'])
pass_math_count = pass_math_grouped['Student ID'].nunique()

#filter and groupby school name reading scores above 70
pass_reading_data = school_data_complete.loc[school_data_complete["reading_score"] >= 70, :]
pass_reading_grouped = pass_reading_data.groupby(['school_name'])
pass_reading_count = pass_reading_grouped['Student ID'].nunique()

#Create Summary Table as a DataFrame
school_summary_df = pd.DataFrame({'School Type': school_type,
                             'Total Students' : total_students.map("{:,}".format),
                             'Total School Budget' : school_budget.map("${:,}".format),
                             'Per Student Budget' : (school_budget/total_students).map("${:.0f}".format),
                             'Average Math Score' : avg_math.map("{:.3f}%".format),
                             'Average Reading Score' : avg_reading.map("{:.3f}%".format),
                             '% Passing Math' : ((pass_math_count/total_students)*100).map("{:.3f}%".format),
                             '% Passing Reading' : ((pass_reading_count/total_students)*100).map("{:.3f}%".format),
                             'Overall Passing Rate': ((((pass_math_count/total_students)+(pass_reading_count/total_students))/2)*100).map("{:.3f}%".format)})

school_summary_df.index.names = ['']
```

```python
# Top 5 performing Schools (by Passing Rate)

top_performing_schools = school_summary_df.sort_values(by = 'Overall Passing Rate', ascending=False)
top_performing_schools.head(5)
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Creat clean dataFrame (no formatting)
bins_summary_df = pd.DataFrame({'School Type': school_type,
                             'Total Students' : total_students,
                             'Total School Budget' : school_budget,
                             'Per Student Budget' : (school_budget/total_students),
                             'Average Math Score' : avg_math,
                             'Average Reading Score' : avg_reading,
                             '% Passing Math' : ((pass_math_count/total_students)*100),
                             '% Passing Reading' : ((pass_reading_count/total_students)*100),
                             'Overall Passing Rate': ((((pass_math_count/total_students)+(pass_reading_count/total_students))/2)*100)})
```

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
labels = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
#Add column Spending Ranges (Per Student) to bins summary df, assing lables based on bins
bins_summary_df['Spending Ranges (Per Student)'] = pd.cut(bins_summary_df["Per Student Budget"], spending_bins, labels=labels)

#groupby Spending Ranges (Per Student)
summary_data_spend = bins_summary_df.groupby("Spending Ranges (Per Student)").mean()

#create dataFrame
summary_data_spend = summary_data_spend[['Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', 'Overall Passing Rate']]

#Format
summary_data_spend = summary_data_spend.astype(float).applymap("{:.3f}%".format)

summary_data_spend
```
